```python
import sys
from pathlib import Path
import cv2
import time

ROOT = Path.cwd().resolve().parent.parent          # si ton notebook est dans /notebooks
# si besoin : ROOT = Path.cwd().resolve().parent  (selon où tu lances)
sys.path.append(str(ROOT / "src"))

from bounce_rl.env.pingpong_env import PingPongEnv
```

```python
print("Démarrage de la simulation PingPongEnv (affichage 30 FPS réel)...")

env = PingPongEnv(
    xml_path="../../assets/mjcf/so101_new_calib copy.xml",
    camera_name="gripper_cam",
    width=320,
    height=240,
    camera_fps=30,   # 30 FPS simulé
    frame_skip=100,
)

DISPLAY_FPS = 30
display_period = 1.0 / DISPLAY_FPS

try:
    obs, info = env.reset()

    print("Observation image shape:", obs["image"].shape)
    print("Observation state shape:", obs["state"].shape)

    done = False
    last_display_time = time.time()

    while not done:

        action = env.action_space.sample()
        obs, reward, terminated, truncated, info = env.step(action)

        # -------- Affichage 30 FPS réel --------
        now = time.time()
        elapsed = now - last_display_time

        if elapsed >= display_period:

            cv2.imshow("Camera View", obs["image"])

            if cv2.waitKey(1) & 0xFF == ord("q"):
                break

            last_display_time = now

        # Synchronisation pour maintenir 30 FPS réel
        sleep_time = display_period - (time.time() - last_display_time)
        if sleep_time > 0:
            time.sleep(sleep_time)

        done = terminated or truncated

except KeyboardInterrupt:
    print("\nCtrl+C détecté. Fermeture propre...")

finally:
    env.close()
    cv2.destroyAllWindows()
    print("Simulation terminée.")
```
